```python
fig= plt.figure (figsize=(10,5))
sns.boxplot (data=df, x= df['price'])
plt.show()
```

```python
fig= plt.figure (figsize=(10,5))
sns.boxplot (data=df, x= df['number_of_reviews'])
plt.show()
```

```python
fig= plt.figure (figsize=(10,5))
sns.boxplot (data=df, x= df['value_price'])
plt.show()
```

```python
#frecuencia absoluta de la col categorica, 'category'
df_categoria= df.groupby('category').count().reset_index()
df_categoria
```

```python
plt.figure(figsize=(8,8))
fig= sns.histplot(x = 'category', data=df_categoria)
plt.show()
```

```python
#frecuencia col cuantitativa
df_precios= df.groupby('price').count().reset_index()
df_precios
```

```python
plt.figure(figsize=(15,8))
fig= sns.barplot (x = 'price', y= 'id', data=df_precios)
plt.xticks(rotation = 45)
plt.show()
```

```python
#son tantos que hacemos una seleccion y representamos la tabla de frecuencias de aquellos precios que estan entre los 5 y los 20 
#es decir, de entre los precios que estan entre 15 y 20, veremos cuales son los que mÃ¡s se repiten
tabla = (df_precios[(df['price']>5) & (df['price']<20)])
tabla.head(2)
```

```python
plt.figure(figsize=(15,8))
fig= sns.barplot (x = 'price', y= 'id', data=tabla)
plt.show()
```
