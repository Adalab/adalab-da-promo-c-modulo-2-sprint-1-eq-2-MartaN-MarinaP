```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv ('attacks_completo.csv')
```

Ayer creamos un par de gráficas para USA y España, donde visualizamos el número de ataques en cada país a lo largo de los años y el tipo de los ataques. Hoy tendréis que hacer dos gráficas con subplots:
- En la primera debéis mostrar los tipos ataques en USA y España, cada subplot corresponderá a un país. Debéis:
* Poner nombre a los ejes de cada uno de los subplots
* Poner un títulos a cada uno de los subplots
* Poner el color de las líneas en negro.
* Establecer el tamaño de la gráfica a 15 x 5
* Ponerle marcadores a las gráficas
* No incluir leyenda


```python
#copiamos el codigo del anterior ejercicio para insertar las gráficas
#primero para USA
df_usa_2000= df[(df['country']=='USA') & (df['year']>2000)]
```

```python
df_tipo= df_usa_2000['type'].value_counts()
df_tipo
```

```python
x=df_tipo.index
y=df_tipo
```

```python
#y ahora para España
df_spain= df[(df['country']=='SPAIN') & (df["year"] >2000)]
```

```python
df_tipo_spain = pd.Series(df_spain['type'].value_counts())
```

```python
df_tipo_spain
```

```python
x1= df_tipo_spain.index
y1=df_tipo_spain
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].plot (x,y, marker = 'v', linewidth = 3)
ax[0].set_ylabel('nº ataques')
ax[0].set_xlabel('tipo de ataques')
ax[0].grid()
ax[0].set_title("Ataques en USA")

ax[1].plot (x1,y1, marker = 'h', linewidth = 3)
ax[1].set_ylabel('nº ataques')
ax[1].set_xlabel('tipo de ataques')
ax[1].grid()
ax[1].set_title("Ataques en España")

plt.show()
```


- En la segunda debéis mostrar los ataques por año para cada uno de los países.
